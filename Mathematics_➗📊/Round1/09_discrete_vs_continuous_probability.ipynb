## **ğŸ“Œ What is a Probability Distribution?**

âœ… A **probability distribution** describes how probabilities are assigned to different outcomes of a random variable.  
âœ… It helps model **real-world random processes** like rolling dice, stock prices, or measuring temperatures.

### **ğŸ”„ Two Main Types:**

| Type | Definition | Example |
|------|------------|---------|
| **Discrete Distribution** ğŸ² | Probability assigned to **countable** outcomes (finite or infinite but countable). | Rolling a die, flipping a coin, number of customers in a store. |
| **Continuous Distribution** ğŸ“ | Probability assigned over an **interval** of infinite possible values. | Heights of people, temperatures, time spent on a website. |

## **ğŸ“Š Key Differences: Discrete vs. Continuous**

| Feature | Discrete Distribution ğŸ² | Continuous Distribution ğŸ“ |
|---------|-------------------------|-------------------------|
| **Definition** | Deals with **countable** values | Deals with **infinite measurable** values |
| **Examples** | Dice rolls ğŸ², number of students in a class ğŸ«, shoe sizes ğŸ‘Ÿ | Heights ğŸ“, weights âš–ï¸, time ğŸ•’ |

## **ğŸ”¢ Example 1: Discrete Probability Distribution (Rolling a Die ğŸ²)**

ğŸ“Œ **Consider rolling a fair 6-sided die:**  
- Possible outcomes: $X = \{1, 2, 3, 4, 5, 6\}$.  
- Each outcome has an equal probability:  
$$P(X = x) = \frac{1}{6}, \quad \text{for } x = 1,2,3,4,5,6$$

ğŸ“Œ **Probability Table:**

| Outcome $X$ | Probability $P(X)$ |
|-------------|--------------------|
| 1           | $\frac{1}{6}$     |
| 2           | $\frac{1}{6}$     |
| 3           | $\frac{1}{6}$     |
| 4           | $\frac{1}{6}$     |
| 5           | $\frac{1}{6}$     |
| 6           | $\frac{1}{6}$     |

âœ… **Observation:**  
The probabilities sum to 1:  
$$P(1) + P(2) + P(3) + P(4) + P(5) + P(6) = 1$$

## **ğŸ”¢ Example 2: Continuous Probability Distribution (Heights of People ğŸ“)**

ğŸ“Œ **Consider the height of adults in cm:**  
- Heights can take **any value** (e.g., **170.2 cm, 171.85 cm, 172.346 cm**).  
- The probability of an **exact height (e.g., 170 cm) is 0**! Instead, we measure probabilities **over a range**.  

## **ğŸ“Œ Probability Calculation:**

To find the probability that a person is **between 170 cm and 180 cm**, we compute:  
$$P(170 \leq X \leq 180) = \int_{170}^{180} f(x) \; dx$$  
where $f(x)$ is the **probability density function (PDF)**.

âœ… **Key Fact:**  
The **area under the curve** represents probability!

## **ğŸ› ï¸ Python Code: Visualizing Discrete vs. Continuous Distributions**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Discrete Distribution: Rolling a Die
X_discrete = np.array([1, 2, 3, 4, 5, 6])
P_discrete = np.array([1/6] * 6)

plt.figure(figsize=(10,4))

# Plot Discrete Distribution
plt.subplot(1,2,1)
plt.bar(X_discrete, P_discrete, color='blue', alpha=0.7)
plt.xlabel("Outcome")
plt.ylabel("Probability")
plt.title("Discrete Distribution: Rolling a Die")

# Continuous Distribution: Normal Curve (Heights)
X_continuous = np.linspace(150, 200, 100)
P_continuous = stats.norm.pdf(X_continuous, loc=175, scale=10)

# Plot Continuous Distribution
plt.subplot(1,2,2)
plt.plot(X_continuous, P_continuous, color='red')
plt.fill_between(X_continuous, P_continuous, alpha=0.3, color='red')
plt.xlabel("Height (cm)")
plt.ylabel("Density")
plt.title("Continuous Distribution: Heights")

plt.tight_layout()
plt.show()
```

# **Probability Density Function (PDF) & Applications in AI/ML ğŸ¤–ğŸ“Š**

## **ğŸ“Œ Quick Overview: Probability Density Function (PDF)**

âœ… The **Probability Density Function (PDF)** describes the **likelihood** of a **continuous random variable** taking a specific range of values.  
âœ… Unlike **discrete probability**, where we calculate $P(X = x)$, in continuous distributions:  
$$P(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx$$  
âœ… The **total area under the PDF curve** equals **1** (since a probability cannot exceed 100%).  
âœ… Used in **normal distributions (Gaussian), exponential, and beta distributions**.

## **ğŸ“Š Example: Normal Distribution PDF**

The **Normal (Gaussian) Distribution** is defined as:  
$$f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}$$  
Where:  
- $\mu$ = **Mean** (center of the distribution).  
- $\sigma^2$ = **Variance** (spread of data).  
- $\sigma$ = **Standard deviation**.

## **ğŸ“Œ Key Property:**

- **68%** of values fall within **1 standard deviation** $(\mu \pm \sigma)$.  
- **95%** within **2 standard deviations** $(\mu \pm 2\sigma)$.  
- **99.7%** within **3 standard deviations** $(\mu \pm 3\sigma)$.

## **ğŸ› ï¸ Python Code: Visualizing a PDF**

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Define the normal distribution
mu, sigma = 0, 1  # Mean and standard deviation
x = np.linspace(-4, 4, 1000)
pdf = stats.norm.pdf(x, mu, sigma)

# Plot the PDF
plt.plot(x, pdf, color='blue', label="Normal Distribution")
plt.fill_between(x, pdf, alpha=0.3, color='blue')
plt.axvline(mu, color='red', linestyle="--", label="Mean (Î¼)")
plt.xlabel("x")
plt.ylabel("Density")
plt.title("Probability Density Function (PDF) of Normal Distribution")
plt.legend()
plt.show()
```

âœ… **Output:** A **bell-shaped curve** centered at $\mu = 0$, showing how probability density is distributed.
